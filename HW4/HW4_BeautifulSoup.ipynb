# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
Country,Region = [],[]
url= 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
rg = requests.get(url)
soup = BeautifulSoup(rg.text,'html.parser')
country = soup.find_all('td',style = 'text-align:left', limit = 50)#top 50 
region = soup.find_all('td',style = 'text-align:center', limit = 50)#top 50 
for i in range(len(country)):
    Country.append(country[i].text)
for i in range(len(region)):
    Region.append(region[i].text)
    
estimate = soup.find_all('td')  
IMF_Estimate = []
for i in range(len(estimate)):
    if i < 51:
        IMF_Estimate.append(estimate[7+(i-1)*8].text)
IMF_Estimate = IMF_Estimate[1:]
IMF_Year = []
for i in range(len(estimate)):
    if i < 51:
        IMF_Year.append(estimate[8+(i-1)*8].text)
IMF_Year = IMF_Year[1:]
United_Nations_Estimate = []
for i in range(len(estimate)):
    if i < 51:
        United_Nations_Estimate.append(estimate[9+(i-1)*8].text)
United_Nations_Estimate = United_Nations_Estimate[1:]

United_Nations_Year = []
for i in range(len(estimate)):
    if i < 51:
        United_Nations_Year.append(estimate[10+(i-1)*8].text)
United_Nations_Year = United_Nations_Year[1:]

World_Bank_Estimate = []
for i in range(len(estimate)):
    if i < 51:
        World_Bank_Estimate.append(estimate[11+(i-1)*8].text)
World_Bank_Estimate = World_Bank_Estimate[1:]

World_Bank_Year,World_Bank_Year1 = [],[]
for i in range(len(estimate)):
    if i < 51:
        World_Bank_Year.append(estimate[12+(i-1)*8].text)
World_Bank_Year = World_Bank_Year[1:]
for i in range(len(World_Bank_Year)):
    World_Bank_Year1.append(World_Bank_Year[i][0:4])
```

```python
# create GDP DataFrame
dict2 = {'Country':Country,'Region':Region,'IMF_Estimate':IMF_Estimate,'IMF_Year':IMF_Year,'United_Nations_Estimate':United_Nations_Estimate,'United_Nations_Year':United_Nations_Year,'World_Bank_Estimate':World_Bank_Estimate,'World_Bank_Year':World_Bank_Year1}
dataframe2 = pd.DataFrame(dict2)
dataframe2
```

```python
# crawler CPI
Country2 = []
url1= 'https://en.wikipedia.org/wiki/Corruption_Perceptions_Index'
rg1 = requests.get(url1)
soup1 = BeautifulSoup(rg1.text,'html.parser')
country2 = soup1.find_all('td',style = 'text-align:left;', limit = 100)
for i in range(len(country2)):
    Country2.append(country2[i].text)
CPI_2020 = []
CPI = soup1.find_all('div',style ='position: absolute; font-weight: bold; width:100%; height: 1.5em; text-align: center; color: black;', limit = 900)
for i in range(len(CPI)):
    if i < 101:
        CPI_2020.append(CPI[0+(i-1)*9].text)
CPI_2020 = CPI_2020[1:]
print(len(CPI_2020))
```

```python
# CPI DataFrame
dict3 = {'Country':Country2,'CPI_2020':CPI_2020}
dataframe3 = pd.DataFrame(dict3)
dataframe3
```

```python
# merge GDP and CPI DataFrame
Dataframe = dataframe2.merge(dataframe3[['Country','CPI_2020']])
Dataframe
```

```python
for i in range(len(Dataframe['CPI_2020'])):
    Dataframe['CPI_2020'][i] = int(Dataframe['CPI_2020'][i])
Dataframe['World_Bank_Estimate'] = Dataframe['World_Bank_Estimate'].str.replace(',','').astype(int)
```

```python
# plot
fig, ax = plt.subplots()
for i in range(len(Dataframe['Country'])):
    if i <= 10:
        ax.annotate(text = Dataframe['Country'][i],xy = (Dataframe['CPI_2020'][i],Dataframe['World_Bank_Estimate'][i]))
p = sns.scatterplot(x = 'CPI_2020', y = 'World_Bank_Estimate',hue ='Region',data = Dataframe)
plt.legend(labels = Dataframe['Region'],loc = 'center right')
```
