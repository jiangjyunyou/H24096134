# (2) The actor generating the highest average revenue? 

```python
import pandas as pd
movie_data=pd.read_csv('D:/IMDB-Movie-Data.csv')#讀取'csv'檔案->'DataFrame'。
#problem 2
movie_data2=movie_data[['Actors','Revenue (Millions)']]#由原有的DataFrame(movie_data)中的'Actors','Revenue (Millions)'指定到新的DataFrame(movie_data2)。
actors_revenue=movie_data2['Actors'].str.split('|',expand=True).stack()#movie_data2中的'Actors'列中的'actor'以'|'區隔開後分成多列。
actors_revenue=actors_revenue.reset_index(level=1,drop=True).rename('actors')#重新設定索引並將'Actors'列重新命名成'actors'。
movie_data2=movie_data2.drop('Actors',axis=1)#刪除'movie_data2'中的'Actors'列。
movie_data2.join(actors_revenue).groupby('actors')[['Revenue (Millions)']].mean().sort_values(['Revenue (Millions)'],ascending=False).head(2)
#Step.1:將'actors'加入'movie_data2'中。#Step.2:計算各'actor'的'Revenue'的平均。#Step.3:各'actor'的'Revenue'的平均由大到小排序，並取出平均值最大者(有兩人並列第一，故 n=2)。
```

# (4) Top‐3 directors who collaborate with the most actors? 

```python
import pandas as pd
movie_data=pd.read_csv('D:/IMDB-Movie-Data.csv')#讀取'csv'檔案->'DataFrame'。
#problem 4
movie_data4=movie_data[['Director','Actors']]#挑出所需資料
x4=movie_data4['Actors'].str.split('|',expand=True).stack()#以'|'分隔各演員
x4=x4.reset_index(level=1,drop=True).rename('actors4')#重新命名columns改為'actors4'
movie_data4=movie_data4.drop('Actors',axis=1)#將未分隔的資料刪去
xx4=movie_data4.join(x4)#將以'|'分隔後的演員資料加入'movie_data4'
xxx4=pd.pivot_table(xx4,values=['actors4'],index=['Director'],aggfunc=lambda x:len(x.unique()))#刪去重複的演員
xxx4.sort_values(['actors4'],ascending=False).head(3)#由大到小排序，並取出合作演員數量前三高之導演
```

# (5) Top‐2 actors playing in the most genres of movies? 

```python
import pandas as pd
movie_data=pd.read_csv('D:/IMDB-Movie-Data.csv')#讀取'csv'檔案->'DataFrame'。
#problem 5
movie_data5=movie_data[['Genre','Actors']]#挑出所需資料
x5=movie_data5['Actors'].str.split('|',expand=True).stack()#以'|'分隔各演員
x5=x5.reset_index(level=1,drop=True).rename('actors5')#重新命名columns改為'actors5'
movie_data5=movie_data5.drop('Actors',axis=1)#將未分隔的資料刪去
xx5=movie_data5.join(x5)#將以'|'分隔後的演員資料加入'movie_data5'
xxx5=xx5['Genre'].str.split('|',expand=True).stack()#以'|'分隔各電影種類
xxx5=xxx5.reset_index(level=1,drop=True).rename('Genres5')#重新命名columns改為'Genres5'
xx5=xx5.drop('Genre',axis=1)#將未分隔的資料刪去
x4_5=xx5.join(xxx5)#將以'|'分隔後的電影種類資料資料加入'xx5'
x4_5['actors5']=x4_5['actors5'].str.strip()#刪去資料中的空格
x4_5=pd.pivot_table(x4_5,values=['Genres5'],index=['actors5'],aggfunc=lambda x:len(x.unique()))#刪除重複的電影種類
x4_5.sort_values(['Genres5'],ascending=False).head(6)#輸出出演電影種類前二高之演員
```

# (6)Top actors whose movies lead to the largest maximum gap of years?

```python
import pandas as pd
movie_data=pd.read_csv('D:/IMDB-Movie-Data.csv')#讀取'csv'檔案->'DataFrame'。
#problem 6
movie_data6=movie_data[['Actors','Year']]#挑出所需資料
x6=movie_data6['Actors'].str.split('|',expand=True).stack()#以'|'分隔各演員
x6=x6.reset_index(level=1,drop=True).rename('actors6')#重新命名columns改為'actors6'
movie_data6=movie_data6.drop('Actors',axis=1)#將未分隔的資料刪去
xx6=movie_data6.join(x6)#將以'|'分隔後的演員資料加入'movie_data6'
xx6['actors6']=xx6['actors6'].str.strip()#刪去資料中的空格
xxx6=xx6.groupby('actors6')[['Year']].aggregate(['max','min'])#以'actors6'作為index，分群各演員出演電影的年份的最大值與最小值
xxx6.insert(xxx6.shape[1],'yeargap',xxx6['Year']['max']-xxx6['Year']['min'])#資料中的最大值扣除嘴最小值，計算'yeargap'
x4_6=xxx6.sort_values(['yeargap'],ascending=False)#找出'yeargap'的最大值為十年
x4_6[x4_6['yeargap']==10]#列出'yeargap'=10年的資料
```
